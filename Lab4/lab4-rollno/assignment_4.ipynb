# KMeans Algorithm

````python
class Kmeans:
    """
    Implementing Kmeans algorithm.

    You may choose to define additional helper function within this class, however don't change the protype of `fit` and `predict` classes.   
    """

    def __init__(self, n_clusters, random_state):
        """
        Implement of kMeans algorithm

        Args:
        n_clusters: int - no. of clusters
        random_state: int - changing this should change initial cluster centers
        """

        self.n_clusters = n_clusters
        self.random_state = random_state

    
    def fit(self, X):
        """
        Implement of kMeans algorithm and stores cluster centers as class object

        Args:
        X : numpy array of shape (n_samples, 2)
        
        Returns:
        Nothing
        """
        ## TODO
        np.random.seed(self.random_state)
        self.cluster_centers = X[np.random.choice(range(X.shape[0]), replace = False, size = self.n_clusters), :]
        
        n_samples = X.shape[0]
        n_dim = X.shape[1]
        old_P = np.zeros(shape=(self.n_clusters, n_samples), dtype=bool)
        
        while(True):
            new_i = np.linalg.norm(np.broadcast_to(X, (self.n_clusters,n_samples,n_dim)).transpose((1,0,2)) - self.cluster_centers, ord=2, axis=2).argmin(axis=1)
            new_P = np.zeros_like(old_P, dtype=bool)
            new_P[new_i, np.arange(n_samples)] = True
            for clus in range(self.n_clusters):
                self.cluster_centers[clus, :] = np.sum(X[new_P[clus]], axis=0)/np.sum(new_P[clus])

            if np.array_equal(new_P, old_P):
                break
            old_P = new_P
        ## END TODO
    
    def predict(self, X):
        """
        Uses the stored cluster centers from fit to predict labels. 
        Args:
        X : numpy array of shape (n_samples, 2)
        
        Returns:
        Y: Labels for each of the data point in X. np
        """
        Y = None
        ## TODO
        n_samples = X.shape[0]
        n_dim = X.shape[1]
        distances = np.linalg.norm(np.broadcast_to(X, (self.n_clusters,n_samples,n_dim)).transpose((1,0,2)) - self.cluster_centers, ord=2, axis=2)
        
        Y = np.argmin(distances, axis=1)
        ## END TODO
        assert Y.shape == (X.shape[0],), "Return Y in expected format"
        return Y
````

# Simple Kernel Design

````python
class Kmeans_Kernel:
    """
    Implementing Kmeans algorithm.

    You may choose to define additional helper function within this class, however don't change the protype of `fit` and `predict` classes.   
    """

    def __init__(self, n_clusters, random_state):
        """
        Implement of kMeans algorithm 

        Args:
        n_clusters: int - no. of clusters
        random_state: int - changing this should change initial cluster centers
        """

        self.n_clusters = n_clusters
        self.random_state = random_state

    def make_zero_centered(self, X):
        """
        Returns a zero-centered concentric circles Dataset

        Args:
        X : numpy array of shape (n_samples, 2)
        
        Returns:
        X : numpy array of shape (n_samples, 2)
        """
        ## TODO
        X -= np.mean(X, axis=0)
        ## END TODO

        return X
    
    def kernel(self, x, y, gamma=5): # custom made kernel
            n = x.shape[0]
            x1 = np.linalg.norm(x, axis=1)
            y1 = np.linalg.norm(y, axis=1)
            # takes in two n*2 length vectors -> n*n matrix with pairwise kernel similarity
#             return (np.broadcast_to(x1, (n, n)).T - y1).T**2
#             return np.exp(-gamma*(np.broadcast_to(x1, (n, n)).T - y1).T**2) # very fast convergence
#             return np.abs((np.broadcast_to(x1, (n, n)).T - y1).T)
#             return np.outer(x1, y1)
            return np.exp(gamma*np.add.outer(x1, y1))
    
    def fit(self, X):
        """
        Implement of kMeans algorithm along with using a suitable kernel and stores cluster centers as class object

        Args:
        X : numpy array of shape (n_samples, 2)
        
        Returns:
        Nothing
        """
        ## TODO
        np.random.seed(self.random_state)
        
        X1 = self.make_zero_centered(X)
        
        self.cluster_centers = np.zeros((self.n_clusters,))
        n_samples = X1.shape[0]
        n_dim = X1.shape[1]
        
        old_y = np.zeros((n_samples,))
        max_iter = 5000
        
        y = np.random.randint(low=0, high=self.n_clusters, size=n_samples)
        K = self.kernel(X1, X1)
        
        for fewf in range(max_iter):
            obj = np.tile(np.diag(K).reshape((-1, 1)), self.n_clusters)
            N_c = np.bincount(y, minlength=self.n_clusters)
            for c in range(self.n_clusters):
                obj[:, c] -= 2 * np.sum((K)[:, y == c], axis=1) / N_c[c]
                obj[:, c] += np.sum((K)[y == c][:, y == c]) / (N_c[c] ** 2)
            y = np.argmin(obj, axis=1)
            if np.array_equal(old_y, y):
                break
            else:
                old_y = y
        
        for c in range(self.n_clusters):
            self.cluster_centers[c] = np.sum(np.linalg.norm(X1, axis=1)[y == c]) / np.sum(y == c) # cluster centers in kernel space
        ## END TODO
    
    def predict(self, X):
        """
        Uses the stored cluster centers from fit to predict labels using the kernel. 
        Args:
        X : numpy array of shape (n_samples, 2)
        
        Returns:
        Y: Labels for each of the data point in X.
        """
        ## TODO
        X1 = self.make_zero_centered(X)
        
        distances = np.square(np.subtract.outer(np.linalg.norm(X1, axis=1), self.cluster_centers))
        Y1 = np.argmin(distances, axis=1)
        
        assert Y1.shape == (X.shape[0],), "Return Y in expected format"
        return Y1
        ## END TODO

X = np.load('./data/train_X_3.npy')
Y = np.array([np.load('./data/train_Y_3.npy')]).T
Y = np.squeeze(Y)

plt.scatter(X[:,0],X[:,1],c=Y)
plt.figure()

km = Kmeans_Kernel(2,123)
km.fit(X)
y_pred = km.predict(X)
plt.scatter(X[:,0],X[:,1],c=y_pred)
````
